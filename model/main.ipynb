```python
#use one where bed or bath =/= 0

df3 = df1.copy()
df3['price_per_sqft'] = df3['price']/df3['Total_Area']
df3.head()
```

```python
len(df3.location.unique())
```

```python
#high dimensionality problem

#how to solve 
#strip data first
df3.location = df3.location.apply(lambda x: x.strip())
location_stats = df3.groupby('location')['location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=30])
```

```python
location_stats_less_than_25 = location_stats[location_stats<=25]
location_stats_less_than_25
```

```python
df3.location = df3.location.apply(lambda x: 'other' if x in location_stats_less_than_25 else x)
len(df3.location.unique())
```

```python
df4 = df3[~(df3.Total_Area/df3.bedrooms<300)]
```

```python
df4.shape
```

```python
df4.price_per_sqft.describe()
```

```python
#remove outliers
#filter anything that is beyond standard deviation

def removeOutliersPricePerArea(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st))&(subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

df5 = removeOutliersPricePerArea(df4)
df5.shape
```

```python
def createScatterPlot(df, location):
    bedrooms_2 = df[(df.location==location)&(df.bedrooms==2)]
    bedrooms_3 = df[(df.location==location)&(df.bedrooms==3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bedrooms_2.Total_Area, bedrooms_2.price, color='blue', label="2 Bedrooms", s=50)
    plt.scatter(bedrooms_3.Total_Area, bedrooms_3.price, color='green', label="3 Bedrooms", s=50)
```
